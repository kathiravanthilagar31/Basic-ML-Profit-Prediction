```python
#pandas is a library which contains functions related to the database
import pandas as pd
```

```python
#Dataset file is assigned to dataset variable 
#pandas used to load the dataset
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
import matplotlib.pyplot as pyp
```

```python
#from the dataset, "yearsExperience" column is assigned to the variable called independent
#which is our Input
#when assigning colum use [[]].
independent=dataset[["YearsExperience"]]
#from the dataset, "Salary" column is assigned to the varibale called dependent
#which is our Output
dependent=dataset[["Salary"]]
dependent
```

```python
pyp.plot(independent,dependent)
```

```python
pyp.scatter(independent,dependent)
pyp.xlabel('YearsExperience',fontsize=20)
pyp.ylabel('Salary',fontsize=20)
pyp.show()
```

```python
independent
```

```python
#sklearn is a library which contains algorithms for machine learning
#model_selection is inside sklearn library, and it contains train_test_split function
from sklearn.model_selection import train_test_split

#train_test_split function works to split our I/P and O/P datas as train and test respectively
#we are assigning variable name for splitted data created from split function
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) 
#test_size is 30 percent of data
```

```python
#from linear model in the sklearn library, we are importing LinearRegression Class
#which perform learning from our dataset to create our model
from sklearn.linear_model import LinearRegression

#we are assigning variable name for the LinearRegression Class = algorithm
regressor=LinearRegression()

#running the function fit from the LinearRegression Class
#the inputs are our training dataset
regressor.fit(X_train,Y_train)
```

```python
#To print weight/slope calaculated in our model through learning
#coef_ is function inside linearRegression Class through which weight/slope were calculated.
weight=regressor.coef_
weight
```

```python
#To print bias/Intercept/origin calculated in our model through learning
#intercept_ is the fuction inside LinearRegression class through which Bias calculated.

bias=regressor.intercept_
bias
```

```python
#Now the model is ready and the purpose is to predict.
#predict is the function inside LinearRegression, used to predict the output.
#It needs Input to perform prediction and we save the output in Y_pred variable
Y_pred=regressor.predict(X_test)
```

```python
#Now, we need to evaluate our prediction
#R square is used for evaluation.
#from metrics in sklearn library, we import r2_score function which perform the evaluation
from sklearn.metrics import r2_score

#R square needs our Actual output and Predicted output to evaluate this model.
#Y_test is our Actual output
#Y_pred is our predicted output from Y_test input
#we save the r2_score function output as r_score
r_score=r2_score(Y_test,Y_pred)
```

```python
r_score
#nearly to 1 in rSquare, best model
```

```python
#To save the best model we created, we need pickle library
import pickle
filename="finalized_model_linear.sav" #sav is extension for pickle
```

```python
#dump is function to save the file.
#we save our model "regressor" to the the filename we gave.
#wb is write binary
pickle.dump(regressor,open(filename,'wb'))
```

```python
#load is the function to load the save model.
#rb is read binary
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))

#Now loaded, to predict, we use the function in our LinearRegression, Predict
# when giving single input, use double closed paranthesis
result=loaded_model.predict([[int(input("Enter your years of Exp:"))]])
```
